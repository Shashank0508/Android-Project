```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers,models,regularizers
import matplotlib.pyplot as plt

np.random.seed(42)
tf.random.set_seed(42)

(x_train,_),(x_test,_)=tf.keras.datasets.mnist.load_data()
x_train=x_train.astype('float32')/255.0
x_test=x_test.astype('float32')/255.0
x_train=x_train.reshape(-1,784)
x_test=x_test.reshape(-1,784)

def corrupt_zero_flipping(data,q):
  mask=np.random.binomial(1,1-q,size=data.shape)
  return data*mask

def corrupt_gaussian_noise(data,sigma):
  noise=np.random.normal(0,sigma,size=data.shape)
  return np.clip(data+noise,0,1)

input_dim=784
hidden_dim=100
l2_lambda=0.01
epochs=20
batch_size=128
q_zero=0.25
sigma_gaussian=0.3


zero_flipping_ae=models.Sequential([
    layers.Input(shape=(input_dim,)),
    layers.Dense(hidden_dim,activation='sigmoid',kernel_regularizer=regularizers.l2(l2_lambda)),
    layers.Dense(input_dim,activation='sigmoid')
])
zero_flipping_ae.compile(optimizer='adam',loss='binary_crossentropy')

x_train_zero_corrupted=corrupt_zero_flipping(x_train,q=q_zero)
zero_flipping_ae.fit(x_train_zero_corrupted,x_train,epochs=epochs,batch_size=batch_size,validation_split=0.1,verbose=1)

gaussian_ae=models.Sequential([
    layers.Input(shape=(input_dim,)),
    layers.Dense(hidden_dim,activation='sigmoid',kernel_regularizer=regularizers.l2(l2_lambda)),
    layers.Dense(input_dim,activation='sigmoid')
])
gaussian_ae.compile(optimizer='adam',loss='binary_crossentropy')

x_train_gaussian_corrupted=corrupt_gaussian_noise(x_train,sigma=sigma_gaussian)
gaussian_ae.fit(x_train_gaussian_corrupted,x_train,epochs=epochs,batch_size=batch_size,validation_split=0.1,verbose=1)
k=10
W_zero=zero_flipping_ae.layers[0].get_weights()[0]
W_gaussian=gaussian_ae.layers[0].get_weights()[0]
x_stars_zero=np.zeros((k,784))
x_stars_gaussian=np.zeros((k,784))

for j in range(k):
  W_j_zero=W_zero[:,j]
  x_star_zero=W_j_zero/np.sqrt(np.sum(W_j_zero**2))
  x_stars_zero[j]=x_star_zero

  W_j_gaussian=W_gaussian[:,j]
  x_star_gaussian=W_j_gaussian/np.sqrt(np.sum(W_j_gaussian**2))
  x_stars_gaussian[j]=x_star_gaussian

#fig,axes=plt.subplots(2,5,figsize=(12,5))
# Plot the learned features
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i in range(k):
    ax = axes[0, i]
    ax.imshow(x_stars_zero[i].reshape(28, 28), cmap='gray')
    ax.axis('off')
    ax.set_title(f'Zero-Flip {i+1}')

    ax = axes[1, i]
    ax.imshow(x_stars_gaussian[i].reshape(28, 28), cmap='gray')
    ax.axis('off')
    ax.set_title(f'Gaussian {i+1}')

plt.tight_layout()
plt.show()


```
